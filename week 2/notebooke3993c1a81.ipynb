```python
# handling missing values with ffill and bfill
train_df_ffill = train_df.fillna(method='ffill', axis=0)
train_df_bfill = train_df_ffill.fillna(method='bfill', axis=0)
```

```python
train_df_bfill['class'].value_counts()
```

```python
#splitting features and target
X_train = train_df_bfill.drop(columns=['ID', 'class'])
y_train = train_df_bfill['class']
```

```python
X_train
```

```python
y_train
```

```python
#preprocessing target variable: class
lencoder = LabelEncoder()
y_train_encoded = lencoder.fit_transform(y_train)
```

```python
# standardize the columns: feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
#training the model
model = LogisticRegression(multi_class='multinomial', max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train_encoded)
```

```python
#evaluating model on training set
train_acc = model.score(X_train_scaled, y_train_encoded)
print('Training Accuracy: ', train_acc)
```

## Testing the model accuracy on testing dataset

```python
y_test_pred_encoded = model.predict(X_test_scaled)
```

```python
y_test_pred = lencoder.inverse_transform(y_test_pred_encoded)
```
