```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
train_df = pd.read_csv('/kaggle/input/summer-analytics-mid-hackathon/hacktrain.csv')
train_df
```

```python
train_df.shape
```

```python
train_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
train_df.isnull().sum()
```

```python
train_df.dtypes
```

```python
ndvi_cols = [col for col in train_df.columns if 'N' in col]
ndvi_cols
```

```python
# filling the missing values using interpolate
train_df[ndvi_cols] = train_df[ndvi_cols].interpolate(method='linear', axis=1, limit_direction='both')
```

```python
train_df.isnull().sum()
```

```python
train_df.dtypes
```

```python
#creating temporal features

#mean ndvi
train_df['ndvi_mean'] = train_df[ndvi_cols].mean(axis=1)
#standard deviation ndvi
train_df['ndvi_std'] = train_df[ndvi_cols].std(axis=1)
# maximum ndvi
train_df['ndvi_max'] = train_df[ndvi_cols].max(axis=1)
#minimum ndvi
train_df['ndvi_min'] = train_df[ndvi_cols].min(axis=1)
#ndvi range
train_df['ndvi_range'] = train_df['ndvi_max'] - train_df['ndvi_min']
```

```python
#assuming earliest is the first column, latest is the last
train_df['ndvi_trend'] = train_df[ndvi_cols[-1]] - train_df[ndvi_cols[0]]
train_df['ndvi_slope'] = train_df['ndvi_trend'] / (len(ndvi_cols) - 1)
```

```python
train_df['ndvi_ratio'] = train_df['ndvi_max'] / (np.abs(train_df['ndvi_min']) + 1e-5)
```

```python
train_df
```

```python
#preparing the data
X_train = train_df.drop(columns=['ID', 'class'])
y_train = train_df['class']
```

```python
# encode target variable
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)

# scale the temporal features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# train the logistic regression model
model = LogisticRegression(max_iter=1000, multi_class='multinomial',class_weight='balanced', C=10, solver='lbfgs')
model.fit(X_train_scaled, y_train_encoded)
```

```python
# evaluating model accuracy on training data
train_acc = model.score(X_train_scaled, y_train_encoded)
print('Training Accuracy', train_acc)
```
